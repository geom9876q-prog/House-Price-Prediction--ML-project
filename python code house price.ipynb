#importing the libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train=pd.read_csv('/kaggle/input/house-price-prediction/train.csv')
df_test=pd.read_csv('/kaggle/input/house-price-prediction/test.csv')
```

#info of the tables 

```python
df_train.info()
df_train.head()
df_train.isnull().sum().sort_values(ascending=False)



```

#clean dataset (numerical)

```python
df_train = df_train.fillna(df_train.median(numeric_only=True))
df_test = df_test.fillna(df_test.median(numeric_only=True))

df_train = df_train.fillna(df_train.mode().iloc[0])
df_test=df_test.fillna(df_test.mode().iloc[0])


df_train.isnull().sum().sort_values(ascending=False)


```

# creating feature and independet variable

```python
X=df_train.iloc[:,:-1]
y=df_train.iloc[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

ENCODING 


```python
X_encoded=pd.get_dummies(X)
test_encoded=pd.get_dummies(df_test)
```

```python
print(X_encoded.shape)
print(test_encoded.shape)
```

```python
X_encoded,test_encoded=X_encoded.align(test_encoded,join='left',axis=1)
test_encoded=test_encoded.fillna(0)
```

```python
print(X_encoded.shape)
print(test_encoded.shape)
```

# SPLITTING FOR TRAIN AND TEST 


```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val=train_test_split(X_encoded,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

# TRAINING THE DATA 

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_val)
np.set_printoptions(precision=2)
```

```python
result=np.column_stack((y_pred,y_val))


print(result)
```

# RMSE VALIDATION 

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print("Validation RMSE:", rmse)
```
